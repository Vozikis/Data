```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Bar plots in pandas

```python
# create a dataset (as a list)
#         young people            old people
data = [ [10, 12, 96, 40],     [90, 20, 55, 15] ]

# labels and age groups
sources = [ 'TV','Newspaper','Internet','Word of mouth' ]
groups  = [ '20-40','50-70' ]

# import into pandas
df = pd.DataFrame(zip(*data),index=sources,columns=groups)
#zip the data. Adding the * takes each elemnt in position i from each array and fits them together
#list(zip(*data)) [(10, 90), (12, 20), (96, 55), (40, 15)]
df
```

```python
# simple bar plot
df.T.plot.bar(figsize=(8,4),ylabel='How often (percent)',
              title='How do people get the news?',fontsize=14)
#figsize=(8,4) making this a little bit larger
plt.show()
```

# Pie chart in matplotlib

```python
# data and labels
preferences = [ 5, 40, 35, 20 ]
iceCreamFlavors = [ 'Fried chicken', 'Chocolate', 'Vanilla', 'Green tea' ]

# explosion factor
explode = [ 0, 0, .05, .15 ]
#how much do you want each piece of the pie to explode meaning how much do you want to grow out of

plt.pie(preferences,labels=iceCreamFlavors,
        explode=explode,shadow=True, autopct='%1.f%%')
# autopct='%1.f%%' corresponds to adding the percentage number on each pie chart here in the graph
plt.show()
```

# Box plots in seaborn

```python
## generate some data

# number of datapoints
N = 200

y1 = np.random.rand(N)
y2 = np.random.randn(N) #gaussian distributed normal numbers

sns.boxplot(data=[y1,y2], palette='Purples') #need to specify that's element inside a list [y1,y2]
sns.stripplot(data=[y1,y2],marker='.',palette='dark:k')# we do this so we don't get points to overlap each other
plt.show()
```

```python
# by importing to pandas

data = np.stack((y1,y2),axis=1)
df = pd.DataFrame(data,columns=['var1','var2'])

df.boxplot();
```

# Histograms with matplotlib

```python
# generate some data

N = 500
data = np.exp( np.random.randn(N)+2 )

plt.figure(figsize=(10,5))
plt.hist(data, bins='fd', density=True)
plt.title('Histogram of data')
plt.xlabel('Data value')
plt.ylabel('Bin count')
plt.show()
```

```python
# using lines instead of bars

d1 = np.random.randn(N)*4
d2 = np.random.randn(N)
d3 = np.random.randn(N)+2

# extract data values
y1,x1 = np.histogram(d1,bins=40)
y2,x2 = np.histogram(d2,bins=40)
y3,x3 = np.histogram(d3,bins=40)

# convert to proportion
y1 = 100*y1/np.sum(y1)
y2 = 100*y2/np.sum(y2)
y3 = 100*y3/np.sum(y3)

# plot the data
plt.figure(figsize=(10,5))
plt.plot((x1[:-1]+x1[1:])/2,y1,'s-',label='Data 1')
plt.plot((x2[:-1]+x2[1:])/2,y2,'o-',label='Data 2')
plt.plot((x3[:-1]+x3[1:])/2,y3,'o-',label='Data 3')

# make the plot look a bit nicer
plt.xlabel('Data value')
plt.ylabel('Percent of total')
plt.legend()
plt.show()
```
